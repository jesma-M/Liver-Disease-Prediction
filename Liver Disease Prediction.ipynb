```python
def convertgender(x):
    if x=='Male':
        return 0
    else:
        return 1
liver_data['Gender'] =liver_data['Gender'].map(convertgender)       
```

```python
liver_data.head()
```

```python
liver_data.drop_duplicates(inplace=True)
```

```python
sns.boxplot(liver_data.Alamine_Aminotransferase)
```

```python
sns.boxplot(liver_data.Aspartate_Aminotransferase)
```

```python
liver_data.Aspartate_Aminotransferase.sort_values(ascending=False).head()
```

```python
liver_data=liver_data[liver_data.Aspartate_Aminotransferase<=3000]
```

```python
liver_data.shape
```

```python
liver_data.Aspartate_Aminotransferase.sort_values(ascending=False).head()
```

```python
liver_data=liver_data[liver_data.Aspartate_Aminotransferase<=2500]
```

```python
liver_data.shape
```

```python
liver_data.isna().sum()
```

```python
liver_data.dropna(how='any',inplace=True)
```

```python
liver_data.isna().sum()
```

```python
liver_data.shape
```

```python
X=liver_data.drop('Dataset',axis=1)
y=liver_data['Dataset']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import GridSearchCV,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
lr_y_pred=lr.predict(X_test)
lr_train=round(lr.score(X_train,y_train)*100,2)
lr_ac=round(accuracy_score(y_test,lr_y_pred)*100,2)
lr_cm=confusion_matrix(y_test,lr_y_pred)
print("Training Accuracy    :",lr_train ,"%")
print("Model Accuracy Score :",lr_ac ,"%")
lr_cm
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
knn_y_pred=knn.predict(X_test)
knn_train=round(knn.score(X_train,y_train)*100,2)
knn_ac=round(accuracy_score(y_test,knn_y_pred)*100,2)
knn_cm=confusion_matrix(y_test,knn_y_pred)
print("Training Accuracy    :",knn_train ,"%")
print("Model Accuracy Score :",knn_ac ,"%")
knn_cm
```

```python
svc=SVC(probability=True)
parameter={
    'gamma':[0.0001,0.001,0.1,1],
    'C':[0.001,0.1,0.005,0.5,1,10,15,20]
}
grid_search=GridSearchCV(svc,parameter)
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
svc=SVC(C= 0.001, gamma= 0.0001)
svc.fit(X_train,y_train)
```

```python
svc_y_pred=svc.predict(X_test)
svc_train=round(knn.score(X_train,y_train)*100,2)
svc_ac=round(accuracy_score(y_test,svc_y_pred)*100,2)
svc_cm=confusion_matrix(y_test,svc_y_pred)
print("Training Accuracy    :",svc_train ,"%")
print("Model Accuracy Score :",svc_ac ,"%")
svc_cm
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
dtc_y_pred=dtc.predict(X_test)
dtc_train=round(dtc.score(X_train,y_train)*100,2)
dtc_ac=round(accuracy_score(y_test,dtc_y_pred)*100,2)
dtc_cm=confusion_matrix(y_test,dtc_y_pred)
print("Training Accuracy    :",dtc_train ,"%")
print("Model Accuracy Score :",dtc_ac ,"%")
dtc_cm
```

```python
rfc=RandomForestClassifier(criterion='entropy', max_depth=15,)
rfc.fit(X_train,y_train)
```

```python
rfc_y_pred=rfc.predict(X_test)
rfc_train=round(rfc.score(X_train,y_train)*100,2)
rfc_ac=round(accuracy_score(y_test,rfc_y_pred)*100,2)
rfc_cm=confusion_matrix(y_test,rfc_y_pred)
print("Training Accuracy    :",rfc_train ,"%")
print("Model Accuracy Score :",rfc_ac ,"%")
rfc_cm
```

```python
gbc=GradientBoostingClassifier()
parameter={
    'loss':['exponential','deviance'],
    'learning_rate':[0.001,0.1,1,10],
    'n_estimators':[100,150,180,200]
}
grid_search_gbc=GridSearchCV(gbc,parameter,cv=20,n_jobs=-1)
grid_search_gbc.fit(X_train,y_train)
```

```python
grid_search_gbc.best_params_
```

```python
gbc=GradientBoostingClassifier(learning_rate= 0.001, loss= 'exponential', n_estimators=100)
gbc.fit(X_train,y_train)
```

```python
gbc_y_pred=gbc.predict(X_test)
gbc_train=round(gbc.score(X_train,y_train)*100,2)
gbc_ac=round(accuracy_score(y_test,gbc_y_pred)*100,2)
gbc_cm=confusion_matrix(y_test,gbc_y_pred)
print("Training Accuracy    :",gbc_train ,"%")
print("Model Accuracy Score :",gbc_ac ,"%")
gbc_cm
```

```python
models=pd.DataFrame({
    'Model':['Logistic Regression','K-NearestNeighbors','Support Vector Machine','Decision Tree','Random Forest','GradientBoosting'],
    'Score':[lr_train,knn_train,svc_train,dtc_train,rfc_train,gbc_train],
    'Accuracy':[lr_ac,knn_ac,svc_ac,dtc_ac,rfc_ac,gbc_ac]
})
```
